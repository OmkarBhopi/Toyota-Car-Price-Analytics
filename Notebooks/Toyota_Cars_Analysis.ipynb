# Toyota Used Cars Market Analysis

Professional EDA for Portfolio, Power BI and ML pipeline.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Toyota_Cleaned.csv')
df.head()
```

## Executive Summary

```python
print('Rows:',df.shape[0])
print('Columns:',df.shape[1])
df.describe().T
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Price'],kde=True)
plt.title('Price Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Age'],kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x='FuelType')
plt.title('Fuel Type Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='Age',y='Price',hue='FuelType')
plt.title('Price vs Age')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='KM',y='Price',hue='FuelType')
plt.title('KM vs Price')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x='FuelType',y='Price')
plt.title('Price by Fuel Type')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df,x='FuelType',y='Price')
plt.title('Average Price by Fuel Type')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python

top_hp = df['HP'].value_counts().head(10)
plt.figure(figsize=(10,6))
top_hp.sort_values().plot(
    kind='barh'
)

plt.title('Top 10 Horsepower Categories')
plt.xlabel('Count')
plt.ylabel('Horsepower')
plt.show()
```

## Business Insights

1. Vehicle age negatively impacts resale value.
2. High mileage vehicles sell at lower prices.
3. Fuel type influences pricing significantly.
4. Horsepower and engine capacity contribute to premium pricing.
5. Dataset is suitable for price prediction ML models.
6. Findings can support dealership inventory and pricing decisions.
